```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import aniposelib
%matplotlib widget
from skimage.morphology import skeletonize
from skimage.util import invert
from skan import draw
from tqdm import tqdm, trange
```

```python
def get_frame(vidcap, framenum):
    vidcap.set(1, framenum);
    _, frame = vidcap.read();
    return frame
```

```python
def get_bg(vidcap, skip = 300):
    nframes = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    bg = np.median([get_frame(vidcap, i) for i in trange(0, nframes, 300)], axis = 0).astype(np.uint8)
    return bg
```

```python
def get_skeleton(frame, bg, threshold_ = 50, gaussian_blur = 11, num_pts = 25):
    fg = cv2.subtract(bg, frame)
    fg_th = cv2.GaussianBlur(fg, (gaussian_blur,gaussian_blur), 0)
    fg_th[fg_th < threshold_] = 0
    fg_th[fg_th >= threshold_] = 255
    contours, hierarchy = cv2.findContours(image=fg_th[:,:,0], 
                                           mode=cv2.RETR_TREE, 
                                           method=cv2.CHAIN_APPROX_NONE)
    fg_contours_sorted = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
    img_leech = np.zeros(fg.shape[:2], dtype = np.uint8)
    cv2.drawContours(image=img_leech, contours=fg_contours_sorted, contourIdx=-1, color=255, thickness=-1, lineType=cv2.LINE_AA)
    img_leech_skeleton = skeletonize(img_leech).astype(np.uint8)*255
    contours, hierarchy = cv2.findContours(image=img_leech_skeleton, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
    skeleton_contours = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
    img_skeleton_coordinates = np.unique(skeleton_contours[0].reshape(-1, 2), axis = 0)
    skeleton = img_skeleton_coordinates[np.linspace(0, len(img_skeleton_coordinates) - 1, num_pts).astype(np.int16)]
    return skeleton
```

```python
def get_tD_skeleton(vidcap0, vidcap1, framenum, calib):
    frame0 = get_frame(vidcap0, framenum)
    frame1 = get_frame(vidcap1, framenum)
    skeleton0 = get_skeleton(frame0, bg0)
    skeleton1 = get_skeleton(frame1, bg1)
    tD = calib.triangulate(np.array([skeleton0, skeleton1]).astype(np.float64))
    reproj = calib.project(tD)
    return tD, reproj
```

```python
def plot_XY(skeleton3d):
    skeleton3d_visualization = skeleton3d - skeleton3d.mean(axis = 0)
    fig, ax = plt.subplots(figsize = (28,7))
    ax.plot(skeleton3d_visualization[:,0], skeleton3d_visualization[:,1])
    ax.set_xlim([-20, 20])
    ax.set_ylim([5, -5])
    ax.set_xticks([])
    ax.set_yticks([])
    plt.tight_layout()
    fig.canvas.draw()
    img = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8, sep='')
    img  = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))
    img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    plt.close()
    return img
```

```python
vidcap0 = cv2.VideoCapture('F:/videos/Camera0/filming_20240110/20240110_1148_24283678_000.mp4')
vidcap1 = cv2.VideoCapture('F:/videos/Camera1/filming_20240110/20240110_1148_24525955_000.mp4')
calib = aniposelib.cameras.CameraGroup.load('F:/session1/calibration/calibration.toml')
```

```python
bg0 = get_bg(vidcap0)
bg1 = get_bg(vidcap1)
fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(bg0)
ax[1].imshow(bg1)
plt.suptitle('Background Images')
plt.tight_layout()
```

```python
frame0 = get_frame(vidcap0, 1250)
frame1 = get_frame(vidcap1, 1250)

fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(frame0)
ax[1].imshow(frame1)
plt.suptitle('Synchronized Leech Frames')
plt.tight_layout()
```

```python
fg0 = cv2.subtract(bg0, frame0)
fg1 = cv2.subtract(bg1, frame1)
fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(fg0)
ax[1].imshow(fg1)
plt.suptitle('Foreground Images')
plt.tight_layout()
```

```python
fg0_th = cv2.GaussianBlur(fg0, (11,11), 0)
fg0_th[fg0_th < 50] = 0
fg0_th[fg0_th >= 50] = 255

fg1_th = cv2.GaussianBlur(fg1, (11,11), 0)
fg1_th[fg1_th < 50] = 0
fg1_th[fg1_th >= 50] = 255

fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(fg0_th)
ax[1].imshow(fg1_th)
plt.suptitle('Thresholded Images')
plt.tight_layout()
```

```python
contours, hierarchy = cv2.findContours(image=fg0_th[:,:,0], mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
fg0_contours_sorted = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
img0_leech = np.zeros(fg0.shape[:2], dtype = np.uint8)
cv2.drawContours(image=img0_leech, contours=fg0_contours_sorted, contourIdx=-1, color=255, thickness=-1, lineType=cv2.LINE_AA)

contours, hierarchy = cv2.findContours(image=fg1_th[:,:,0], mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
fg1_contours_sorted = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
img1_leech = np.zeros(fg1.shape[:2], dtype = np.uint8)
cv2.drawContours(image=img1_leech, contours=fg1_contours_sorted, contourIdx=-1, color=255, thickness=-1, lineType=cv2.LINE_AA)

fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(img0_leech, cmap = 'gray')
ax[1].imshow(img1_leech, cmap = 'gray')
plt.suptitle('Images with leech contours')
plt.tight_layout()
```

```python
img0_leech_skeleton = skeletonize(img0_leech).astype(np.uint8)*255
img1_leech_skeleton = skeletonize(img1_leech).astype(np.uint8)*255

fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(img0_leech_skeleton, cmap = 'gray')
ax[1].imshow(img1_leech_skeleton, cmap = 'gray')
plt.suptitle('Skeletonized Leech')
plt.tight_layout()
```

```python
contours, hierarchy = cv2.findContours(image=img0_leech_skeleton, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
skeleton_contours = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
img0_skeleton_coordinates = np.unique(skeleton_contours[0].reshape(-1, 2), axis = 0)

contours, hierarchy = cv2.findContours(image=img1_leech_skeleton, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
skeleton_contours = [sorted(contours, key=lambda x: cv2.contourArea(x))[::-1][0]]
img1_skeleton_coordinates = np.unique(skeleton_contours[0].reshape(-1, 2), axis = 0)


fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(frame0)
ax[0].plot(img0_skeleton_coordinates[:,0], img0_skeleton_coordinates[:,1])
ax[1].imshow(frame1)
ax[1].plot(img1_skeleton_coordinates[:,0], img1_skeleton_coordinates[:,1])
plt.suptitle('Leech Images with Skeletonized Overlay')
plt.tight_layout()
```

```python
skeleton0 = img0_skeleton_coordinates[np.linspace(0, len(img0_skeleton_coordinates) - 1, 25).astype(np.int16)]
skeleton1 = img1_skeleton_coordinates[np.linspace(0, len(img1_skeleton_coordinates) - 1, 25).astype(np.int16)]
tD = calib.triangulate(np.array([skeleton0, skeleton1]).astype(np.float64))
reproj = calib.project(tD)
fig,ax = plt.subplots(1, 2, figsize = (8,2))
ax[0].imshow(frame0)
ax[0].plot(reproj[0][:,0], reproj[0][:,1])
ax[1].imshow(frame1)
ax[1].plot(reproj[1][:,0], reproj[1][:,1])
plt.suptitle('3D Skeletonized Leech projected back onto the cameras')
plt.tight_layout()
```
